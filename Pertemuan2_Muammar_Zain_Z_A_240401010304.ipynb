```python
import numpy as np
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(url)
df.head(5)
```

```python
print("QUIZ NOMOR 1")
print("Jumlah Baris: ", df.shape[0])
print("Jumlah Kolom: ", df.shape[1])
```

```python
print("QUIZ NOMOR 2")
print(df.isnull().sum())
```

```python
print("QUIZ NOMOR 3")
jumlah_penumpang = df[df['Survived'] == 1].shape[0]
persen_penumpang = jumlah_penumpang / df.shape[0] * 100
print("Persentase Penumpang yang selamat: ", persen_penumpang, "%")
```

```python
print("QUIZ NOMOR 4")
wk1 = df[(df['Sex'] == "female") & (df['Pclass'] == 1) & (df['Survived'] == 1)]
print("Rata - Rata Usia Penumpang Wanita Kelas 1 yang selamat adalah: ",
      wk1['Age'].mean())
```

```python
print("QUIZ NOMOR 5")
count_pclass = (
    df[df['Survived'] == 1]
    .groupby('Pclass')
    .size()
    .reset_index(name="count")
    .sort_values(by = "count", ascending=False)
)
print(count_pclass)
print("Kelas dengan tingkat keselamatan tertinggi adalah: ", count_pclass.iloc[0]['Pclass'])
```

```python
print("QUIZ NOMOR 6")
print(df.groupby('Pclass')['Survived'].mean())
```
